## Ridge Regression

```python
# Standardization
```

```python
#importing preprocessing modue from sklearn library to standardize data before performing model
from sklearn import preprocessing
```

```python
scaler=preprocessing.StandardScaler().fit(x_train)
```

```python
#transforming x_train variables into scaler variables
x_train_s=scaler.transform(x_train)
```

```python
#transforming x_test variables into scaler variables
x_test_s=scaler.transform(x_test)
```

```python
#Importing ridge model
from sklearn.linear_model import Ridge
```

```python
#Ridge regression model
lm_r=Ridge(alpha=0.5)
```

```python
ridge_mse_train=mean_squared_error(y_train,lm_r.predict(x_train_s))
```

```python
ridge_mse_test=mean_squared_error(y_test,lm_r.predict(x_test_s))
```

```python
#Evaluating model with R_squared_score and Mean_Squared_Error
print("R-squared Score (Train) - Ridge Regression:", ridge_r2_square_train)
print("R-squared Score (Test) - Ridge Regression:", ridge_r2_square_test)
print("mean_squared_error (Train) - Ridge Regression:",ridge_mse_train)
print("mean_squared_error (Test) - Ridge Regression:", ridge_mse_test)
```

## Validation curve

```python
#importing validation curve
from sklearn.model_selection import validation_curve
```

```python
# Define a range of alpha values for Ridge regression
param_range=np.logspace(-2,8,100)
```

```python
param_range
```

```python
# Perform cross-validated validation curve to find the best alpha for Ridge regression
train_score, test_score = validation_curve(Ridge(), x_train_s, y_train,
                                           param_name="alpha",
                                           param_range=param_range,
                                            scoring="r2")
```

```python
print(train_score)
```

```python
print(test_score)
```

```python
# Calculate mean R-squared scores for training and validation sets
train_mean=np.mean(train_score,axis=1)
```

```python
test_mean=np.mean(test_score,axis=1)
```

```python
train_mean
```

```python
max(train_mean)
```

```python
sns.jointplot(x=np.log(param_range),y=test_mean)
```

```python
# Finding the alpha value with the highest validation R-squared
np.where(test_mean==max(test_mean))
```

```python
param_range[26]
```

```python
# Train a Ridge regression model with the best alpha
lm_r_best=Ridge(alpha=param_range[26])
```

```python
lm_r_best.fit(x_train_s,y_train)
```

```python
lm_r_best_train_r2_score=r2_score(y_train,lm_r_best.predict(x_train_s))
```

```python
lm_r_best_test_r2_score=r2_score(y_test,lm_r_best.predict(x_test_s))
```

```python
lm_r_best_test_mse=mean_squared_error(y_test,lm_r_best.predict(x_test_s))
```

```python
lm_r_best_train_mse=mean_squared_error(y_train,lm_r_best.predict(x_train_s))
```

```python
#Evaluating Ridge regression model with the best alpha
print("R-squared Score (Train) - Ridge Regression vc:", lm_r_best_train_r2_score)
print("R-squared Score (Test) - Ridge Regression vc:", lm_r_best_test_r2_score)
print("mean_squared_error (Train) - Ridge Regression vc:",lm_r_best_test_mse)
print("mean_squared_error (Test) - Ridge Regression vc:", lm_r_best_train_mse)
```

```python
mse_train_bagging = mean_squared_error(y_train,y_train_pred_bagging)
```

```python
print("R-squared Score (Train) - Bagging Regressor:", r2_score_bagging_train)
print("R-squared Score (Test) - Bagging Regressor:", r2_score_bagging_test)
print("mean_squared_error (Train) - Bagging Regressor:", mse_train_bagging)
print("mean_squared_error (Test) - Bagging Regressor:", mse_test_bagging)
```
